```python
# استخراج بخش بالینی
clinical_cols = ['معاینات بالینی'] + [f'Unnamed: {i}' for i in range(119, 145)]
df_clinical = df[clinical_cols].copy()

# نام‌گذاری
rename_map = {
    'معاینات بالینی': 'BP_Systolic',
    'Unnamed: 119': 'BP_Diastolic',
    'Unnamed: 120': 'Heart_Rate',
    'Unnamed: 121': 'Respiratory_Rate',
    'Unnamed: 122': 'Temperature',
    'Unnamed: 123': 'SpO2_No_O2',
    'Unnamed: 124': 'SpO2_With_O2',
    'Unnamed: 125': 'Consciousness',
    'Unnamed: 141': 'ICU_Admission',
    'Unnamed: 143': 'Intubation',
    'Unnamed: 132': 'Observations',
    'Unnamed: 133': 'Lung_Auscultation',
}
df_clinical.rename(columns=rename_map, inplace=True)
```

```python
# 4️⃣ حذف ردیف‌هایی که در واقع هدر هستن (مثل ردیف 0 و 1 در نمونه‌ها)
# این ردیف‌ها شامل متن‌هایی مثل "فشار خون سیستولیک" یا "درصد اشباع اکسیژن" هستن
df_clinical = df_clinical[
    ~df_clinical['BP_Systolic'].astype(str).str.contains('فشار خون|سیستولیک', na=True)
]
df_clinical = df_clinical[
    ~df_clinical['SpO2_No_O2'].astype(str).str.contains('اشباع|اکسیژن|درصد', na=True)
]
df_clinical = df_clinical[
    ~df_clinical['Heart_Rate'].astype(str).str.contains('ضربان|قلب', na=True)
]

# ریست ایندکس بعد از حذف
df_clinical.reset_index(drop=True, inplace=True)

print(f"تعداد رکورد بعد از حذف هدرهای اشتباه: {len(df_clinical)}")
```

```python
# 5️⃣ تمیز کردن مقادیر متنی قبل از تبدیل به عدد
numeric_cols = ['BP_Systolic', 'BP_Diastolic', 'Heart_Rate',
                'Respiratory_Rate', 'Temperature',
                'SpO2_No_O2', 'SpO2_With_O2']

for col in numeric_cols:
    df_clinical[col] = (
        df_clinical[col]
        .astype(str)
        .str.strip()
        .str.replace('%', '', regex=False)
        .str.replace(r'\.$', '', regex=True)  # حذف نقطه آخر مثل 37.
        .replace('nan', np.nan)
    )
# 6️⃣ تبدیل به عددی (هر چیزی که نشد → NaN)
df_clinical[numeric_cols] = df_clinical[numeric_cols].apply(pd.to_numeric, errors='coerce')

# 7️⃣ تمیز کردن ستون‌های کتگوریکال
cat_cols = ['Consciousness', 'ICU_Admission', 'Intubation',
            'Observations', 'Lung_Auscultation']

for col in cat_cols:
    df_clinical[col] = (
        df_clinical[col]
        .astype(str)
        .str.strip()
        .str.lower()
        .replace('nan', np.nan)
    )
```

```python
# 8️⃣ پر کردن هوشمند مقادیر گمشده

# اول: عددی‌ها با KNN Imputer (بهترین روش برای داده‌های بالینی)
print("در حال اجرای KNN Imputer برای فیچرهای عددی...")
knn_imputer = KNNImputer(n_neighbors=5, weights='uniform')
df_clinical[numeric_cols] = knn_imputer.fit_transform(df_clinical[numeric_cols])

# دوم: کتگوریکال با شایع‌ترین مقدار (mode)
for col in cat_cols:
    mode_val = df_clinical[col].mode()[0] if not df_clinical[col].mode().empty else 'unknown'
    df_clinical[col].fillna(mode_val, inplace=True)
```

```python
# 9️⃣ نتیجه نهایی
print("\n✅ پیش‌پردازش کامل شد!")
print("شکل نهایی دیتافریم بالینی:", df_clinical.shape)
print("تعداد NaN باقی‌مانده در هر ستون:")
print(df_clinical.isna().sum())

print("\n📊 آمار توصیفی فیچرهای عددی:")
display(df_clinical[numeric_cols].describe())

print("\n👀 نمونه ۱۰ رکورد اول:")
display(df_clinical.head(10))

print("\n🎯 حالا قسمت علائم بالینی دیتاست کاملاً تمیز و آماده است :")
```

```python
# =========================
# Vaccination History (Final)
# =========================

# 1) انتخاب ستون‌ها
vax_cols = ['سابقه واکسیناسیون', 'Unnamed: 100', 'Unnamed: 101']
df_vax_raw = df_raw[vax_cols].copy()

# حذف ردیف‌های متنی
df_vax = df_vax_raw.iloc[3:].reset_index(drop=True)

# 2) تعداد دوز (مهم‌ترین ستون)
df_vax['Vaccine_Doses'] = pd.to_numeric(
    df_vax['Unnamed: 100'],
    errors='coerce'
).fillna(0)

# 3) واکسینه شده یا نه (تعریف درست)
df_vax['Vaccinated'] = (df_vax['Vaccine_Doses'] > 0).astype(int)

# 4) ساده‌سازی نوع واکسن (اختیاری – فقط گزارش)
def simplify_vaccine(x):
    if pd.isna(x):
        return 'Unknown'
    x = str(x).lower()
    if 'سینوفارم' in x or 'sinopharm' in x:
        return 'Sinopharm'
    if 'آستراز' in x or 'astraz' in x:
        return 'AstraZeneca'
    if 'برکت' in x:
        return 'Barkat'
    if 'اسپوتنیک' in x:
        return 'Sputnik'
    return 'Other'

df_vax['Vaccine_Type_Simple'] = df_vax['Unnamed: 101'].apply(simplify_vaccine)

# 5) دیتافریم نهایی برای مدل
df_vax_final = df_vax[['Vaccinated', 'Vaccine_Doses']].copy()

# 6) خروجی تحلیلی
print("📐 شکل دیتافریم Vaccination:", df_vax_final.shape)

print("\n🔍 نمونه داده‌ها:")
display(df_vax_final.head(10))

print("\n📊 آمار واکسیناسیون:")
print("درصد واکسینه‌شده‌ها:", round(df_vax_final['Vaccinated'].mean()*100, 2))

print("\n📊 توزیع تعداد دوز:")
display(df_vax_final['Vaccine_Doses'].value_counts().sort_index())

print("\n📊 توزیع نوع واکسن (گزارشی):")
display(df_vax['Vaccine_Type_Simple'].value_counts())
```

```python
# =========================
# Drop Irrelevant / High-Missing Columns
# =========================

df = df_raw.copy()

# 1) حذف ستون‌های فامیلی و مشاوره (بر اساس نام)
drop_keywords = [
    'فامیلی',
    'خانوادگی',
    'مشاوره',
    'consult',
    'family'
]

cols_to_drop = [
    col for col in df.columns
    if any(key in str(col) for key in drop_keywords)
]

# 2) حذف ستون‌های Unnamed با missing بالا
missing_threshold = 0.9  # 90 درصد

high_missing_cols = [
    col for col in df.columns
    if col.startswith('Unnamed')
    and df[col].isna().mean() > missing_threshold
]

# 3) حذف نهایی
final_drop_cols = list(set(cols_to_drop + high_missing_cols))

df_cleaned = df.drop(columns=final_drop_cols)

# 4) گزارش
print(f"🧹 تعداد ستون‌های حذف‌شده: {len(final_drop_cols)}")
print("\n🗑 نمونه ستون‌های حذف‌شده:")
for c in final_drop_cols[:20]:
    print("-", c)

print("\n📐 شکل دیتافریم بعد از پاکسازی:")
print(df_cleaned.shape)
```
